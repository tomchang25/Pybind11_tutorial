```python
def show_landmarks(image, landmarks):
    """Show image with landmarks"""
    plt.imshow(image)
    plt.scatter(landmarks[:, 0], landmarks[:, 1], s=10, marker='.', c='r')
    plt.pause(0.001)  # pause a bit so that plots are updated
```

```python
rescale = Rescale(256)
crop = RandomCrop(224)
pyrescale = PyRescale(256)
pycrop = PyRandomCrop(224)
```

```python
"""Face Landmarks dataset."""
idx = 0
landmarks_frame = pd.read_csv(csv_file)


images = np.zeros((0,224,224,3))
landmarks = np.zeros((0,68,2))

landmark = landmarks_frame.iloc[idx, 1:]
landmark = np.array([landmark]).astype('double').reshape(-1, 2)

img_name = os.path.join(root_dir, landmarks_frame.iloc[idx, 0])
image = io.imread(img_name).astype('uint8')

T3 = time.time()
image, landmark = rescale(image, landmark)
image,landmark = crop(image, landmark)
T4 = time.time()

show_landmarks(image,landmark)

img_name = os.path.join(root_dir, landmarks_frame.iloc[idx, 0])
image = io.imread(img_name).astype('uint8')
landmark = landmarks_frame.iloc[idx, 1:]
landmark = np.array([landmark]).astype('double').reshape(-1, 2)

T5 = time.time()
image, landmark = pyrescale(image, landmark)
image,landmark = pycrop(image, landmark)
T6 = time.time()
show_landmarks(image,landmark)

print("C++ version take:    ",T4-T3,"\nPython versoin take: ",T6-T5)
```
