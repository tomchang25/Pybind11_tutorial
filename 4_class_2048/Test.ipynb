```python
board = Board()
```

```python
while(1):
    if board.game_over():
        print("~~~GAME OVER~~~")
        
    board.display();
    print("Score: ", board.score)
    print("Step: ", board.step)
    
    ch = input()
    
    if ch == 'w':
        board.move_up()
    elif ch == 's':
        board.move_down()
    elif ch == 'a':
        board.move_left()
    elif ch == 'd':
        board.move_right()

```
