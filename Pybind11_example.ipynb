```c++
#include <pybind11/pybind11.h>
#include <pybind11/stl.h>
#include <pybind11/numpy.h>

// main libarary of xtensor
#include "xtensor/xarray.hpp"

// bind numpy to xtensor
#define FORCE_IMPORT_ARRAY                // numpy C api loading
#include "xtensor-python/pyarray.hpp"     // Numpy bindings

namespace py = pybind11;

using namespace std;

void hello_world(){
    py::print("Hello world! ");
}

string say_hello(string name){
    return name + "說你好.";
}


class Myclass {
public:
    // parameter
    int output_size;
    // init
    Myclass(int output_size) : output_size(output_size) { } 
    
    // function
    vector<int> get(){        
        return vector<int>(this->output_size);
    }
    
    // numpy function
    xt::pyarray<double> sub(xt::pyarray<double> arr,int num){
        return arr - num;
    }
};


PYBIND11_MODULE(test, m) {
  xt::import_numpy();

  m.doc() = "pybind11 example plugin";      // module doc string
  m.def("hello_world",&hello_world);
  m.def("say_hello",&say_hello);
    
    
  py::class_<Myclass>(m, "Myclass")
    .def(py::init<int>())
    .def("get",&Myclass::get)
    .def("sub",&Myclass::sub);

}
```

```c++
# numpy to cv::Mat
cv::Mat numpy_uint8_3c_to_cv_mat(py::array_t<unsigned char>& input) {

    if (input.ndim() != 3)
        throw std::runtime_error("3-channel image must be 3 dims ");

    py::buffer_info buf = input.request();

    cv::Mat mat(buf.shape[0], buf.shape[1], CV_8UC3, (unsigned char*)buf.ptr);

    return mat;
}

# cv:Mat to numpy 
py::array_t<unsigned char> cv_mat_uint8_3c_to_numpy(cv::Mat& input) {

    py::array_t<unsigned char> dst = py::array_t<unsigned char>({ input.rows,input.cols,3}, input.data);
    return dst;
}
```

```c++
struct buffer_info {
    void *ptr;
    ssize_t itemsize;
    std::string format;
    ssize_t ndim;
    std::vector<ssize_t> shape;
    std::vector<ssize_t> strides;
};
```
